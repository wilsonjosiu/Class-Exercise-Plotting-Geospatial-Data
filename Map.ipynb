```python
import folium
from folium.features import DivIcon
import csv
import branca
from geopy.geocoders import Nominatim
```

```python
data = {}
geolocator = Nominatim(user_agent="specify_your_app_name_here")

m = folium.Map(
    location=[48, -102],
    titles='The average temperature of several US cities',
    zoom_start=3
    )
```

```python
with open('city_temperatures.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter = ',', quotechar = '|')
    for row in spamreader:
        if (row[0][0] != '%'):
            city = row[0]
            state = row[1]
            temperature = float(row[2])
            location = geolocator.geocode(city, state)
            data[city, state] = {'temperature': temperature, 'latitude': location.latitude, 'longitude': location.longitude}
```

```python
colorscale = branca.colormap.LinearColormap(colors=[(85,72,193),(242,242,242),(177,1,39)], index=[53,57,68], vmin=50, vmax=70)
```

```python
colorscale
```

```python
for key in data:
    tempo = str(colorscale(data[key]['temperature']))
    T = tempo[0:7]
    folium.CircleMarker(
        location=[data[key]['latitude'], data[key]['longitude']],
        radius=data[key]['temperature']/3,
        fill=True,
        fill_color=T
        ).add_to(m)
    folium.map.Marker([data[key]['latitude'], data[key]['longitude']], icon=DivIcon(icon_size=(150,36), icon_anchor=(10,10), html='<div style"font-size:24pt">%s</div>' % data[key]['temperature'])).add_to(m)
```
